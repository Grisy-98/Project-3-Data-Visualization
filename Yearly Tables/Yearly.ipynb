```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json 
```

```python
mhc_2018 = pd.read_csv("Median House Cost/2018 Median House Cost.csv", index_col=0)
mhc_2019 = pd.read_csv("Median House Cost/2019 Median House Cost.csv", index_col=0)
mhc_2020 = pd.read_csv("Median House Cost/2020 Median House Cost.csv", index_col=0)
mhc_2021 = pd.read_csv("Median House Cost/2021 Median House Cost.csv", index_col=0)
mhc_2022 = pd.read_csv("Median House Cost/2022 Median House Cost.csv", index_col=0)
mhc_2018
```

```python
mmc_2018 = pd.read_csv("Median Monthly Cost/2018_Monthly_Median_Cost.csv", index_col=0)
mmc_2019 = pd.read_csv("Median Monthly Cost/2019_Monthly_Median_Cost.csv", index_col=0)
mmc_2020 = pd.read_csv("Median Monthly Cost/2020_Monthly_Median_Cost.csv", index_col=0)
mmc_2021 = pd.read_csv("Median Monthly Cost/2021_Monthly_Median_Cost.csv", index_col=0)
mmc_2022 = pd.read_csv("Median Monthly Cost/2022_Monthly_Median_Cost.csv", index_col=0)

mmc_2018
```

```python
tenure_2018 = pd.read_csv("Tenure/2018 Tenure.csv", index_col=0)
tenure_2019 = pd.read_csv("Tenure/2019 Tenure.csv", index_col=0)
tenure_2020 = pd.read_csv("Tenure/2020 Tenure.csv", index_col=0)
tenure_2021 = pd.read_csv("Tenure/2021 Tenure.csv", index_col=0)
tenure_2022 = pd.read_csv("Tenure/2022 Tenure.csv", index_col=0)

tenure_2020
```

```python
mcr_2018 = pd.read_csv("Median Contract Rent/2018 Median Contract Rent.csv", index_col=0)
mcr_2019 = pd.read_csv("Median Contract Rent/2019 Median Contract Rent.csv", index_col=0)
mcr_2020 = pd.read_csv("Median Contract Rent/2020 Median Contract Rent.csv", index_col=0)
mcr_2021 = pd.read_csv("Median Contract Rent/2021 Median Contract Rent.csv", index_col=0)
mcr_2022 = pd.read_csv("Median Contract Rent/2022 Median Contract Rent.csv", index_col=0)
mcr_2022
```

```python
merge_2018_1 = pd.merge(mcr_2018, mhc_2018, on='Zip Code')
merge_2018_2 = pd.merge(merge_2018_1, mmc_2018, on='Zip Code')
df_2018 = pd.merge(merge_2018_2, tenure_2018, on='Zip Code')



df_2018 = df_2018.rename(columns={
    "2018 Median Rent" : "2018 Median Contract Rent",
    "2018 Median Cost" : "Median Monthly Cost",
    "2018 Total" : "Total Median Monthly Cost",
    "2018 With Mortgage" : "Median Monthly Cost With Mortgage",
    "2018 Without Mortgage" : "Median Monthly Cost Without Mortgage",
    "2018 Total Units" : "Total Housing Units",
    "2018 Number Owned" : "Housing Units Owned",
    "2018 Number Rented" : "Housing Units Rented",
})

df_2018
```
